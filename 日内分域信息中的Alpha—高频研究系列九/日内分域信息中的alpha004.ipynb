```python
# 定义因子列
factor_columns = ['intraDSVol_byT', 'intraDSRtn_byT', 'intraDSStd_byP', 'intraDSVol_byP', 'intraDSRtn_byV', 'intraDSStd_byV']

# 初始化 Rank IC 结果数据框
ic_results = pd.DataFrame(index=factor_columns, columns=['IC均值', 'IC标准差', 'ICIR', 'T统计量'])

# 计算每个因子的 Rank IC
for factor in factor_columns:
    daily_ic = []
    
    # 强制转换为 float 类型，以避免 dtype 问题
    merged_data[factor] = pd.to_numeric(merged_data[factor], errors='coerce')
    merged_data['daily_return'] = pd.to_numeric(merged_data['daily_return'], errors='coerce')
    
    for day, group in merged_data.groupby('day'):
        # 确保两列都不为空，且至少有两个非 NaN 数据点
        if group[factor].notna().sum() > 1 and group['daily_return'].notna().sum() > 1:
            # 计算 Spearman 相关系数，忽略 NaN
            ic = group[factor].corr(group['daily_return'], method='spearman')
            daily_ic.append(ic)
    
    if len(daily_ic) > 0:  # 确保存在有效数据进行计算
        ic_mean = np.mean(daily_ic)
        ic_std = np.std(daily_ic)
        ic_results.loc[factor, 'IC均值'] = ic_mean
        ic_results.loc[factor, 'IC标准差'] = ic_std
        ic_results.loc[factor, 'ICIR'] = ic_mean / ic_std if ic_std != 0 else np.nan
        ic_results.loc[factor, 'T统计量'] = ic_mean / (ic_std / np.sqrt(len(daily_ic))) if ic_std != 0 else np.nan

# 输出 Rank IC 结果
print("Rank IC 测试结果")
print(ic_results)
```

	1.	多空收益率: 这通常是通过构建多头和空头组合的收益率差计算的。例如，可以将因子值高的股票作为多头，因子值低的股票作为空头组合，计算他们的平均收益差异。
	2.	年化波动率: 通常使用标准差乘以年化因子（如252，用于日频数据）来计算。
	3.	夏普比率: 夏普比率计算公式为：\text{Sharpe Ratio} = \frac{\text{Mean Return} - \text{Risk-Free Rate}}{\text{Standard Deviation}}。在没有无风险利率的情况下，我们可以只用收益率的平均值除以波动率。
	4.	最大回撤: 最大回撤用于衡量策略回撤的程度，即在投资过程中从高峰到低谷的最大跌幅。通常通过累计收益计算。
	5.	多头换手率: 通常是一个交易频率指标，表示多头仓位的变动频率。

```python
import numpy as np
import pandas as pd

# 示例因子数据和每日收益率
# merged_data['factor_values'] = your factor values
# merged_data['daily_return'] = your daily return data

# 分别计算多头和空头组合的收益率
def calculate_long_short_performance(data, factor_column):
    # 假设因子值高的30%是多头，低的30%是空头
    long_stocks = data[data[factor_column] >= data[factor_column].quantile(0.7)]
    short_stocks = data[data[factor_column] <= data[factor_column].quantile(0.3)]

    # 计算多头、空头和多空组合的收益率
    long_return = long_stocks['daily_return'].mean()
    short_return = short_stocks['daily_return'].mean()
    long_short_return = long_return - short_return

    # 年化波动率
    annual_vol = data['daily_return'].std() * np.sqrt(252)
    
    # 夏普比率
    sharpe_ratio = long_short_return / annual_vol
    
    # 计算累计收益和最大回撤
    cumulative_returns = (1 + data['daily_return']).cumprod()
    peak = cumulative_returns.cummax()
    drawdown = (peak - cumulative_returns) / peak
    max_drawdown = drawdown.max()

    return long_short_return, long_return, short_return, annual_vol, sharpe_ratio, max_drawdown

# 定义因子列
factor_columns = ['intraDSVol_byT', 'intraDSRtn_byT', 'intraDSStd_byP', 'intraDSVol_byP', 'intraDSRtn_byV', 'intraDSStd_byV']

# 初始化结果 DataFrame
performance_results = pd.DataFrame(columns=['多空收益率', '多头收益率', '空头收益率', '年化波动率', '夏普比率', '最大回撤'])

# 计算每个因子的多空组合表现
for factor in factor_columns:
    long_short_rtn, long_rtn, short_rtn, annual_vol, sharpe_ratio, max_dd = calculate_long_short_performance(merged_data, factor)
    performance_results.loc[factor] = [long_short_rtn, long_rtn, short_rtn, annual_vol, sharpe_ratio, max_dd]

# 显示结果
performance_results
```

### 3.5

	1.	准备数据：确保数据中存在时间分域特征 (qcutCHR) 以及相关的收益率 (rtn)。
	2.	计算显著性因子：使用公式 intraDS2 = max(|qcutCHRi - avg(qcutCHR)| / (|qcutCHRi| + |avg(qcutCHR)| + epsilon)) * rtn[argmax]。
	3.	选取最大显著性的分域：根据公式，选出显著性最强的分域对应的收益率。

```python
import numpy as np
import pandas as pd

# 使用 'intraDSRtn_byT' 列作为 qcutCHR 代表，计算其平均值
merged_data['avg_intraDSRtn_byT'] = merged_data['intraDSRtn_byT'].mean()

# 防止除以零的问题，定义一个很小的 epsilon 值
epsilon = 1e-5

# 计算 intraDS2
merged_data['intraDS2'] = np.abs(merged_data['intraDSRtn_byT'] - merged_data['avg_intraDSRtn_byT']) / (
    np.abs(merged_data['intraDSRtn_byT']) + np.abs(merged_data['avg_intraDSRtn_byT']) + epsilon
)

# 找到显著性最强的分域（intraDS2 最大值），检查对应的收益率 (rtn_morning) 是否为零
argmax_idx = merged_data['intraDS2'].idxmax()
rtn_max = merged_data.loc[argmax_idx, 'rtn_morning']

# 如果最大显著性的收益率为零，寻找下一个不为零的收益率
if rtn_max == 0:
    # 查找非零的最大收益率
    non_zero_rtn = merged_data[merged_data['rtn_morning'] != 0]
    if len(non_zero_rtn) > 0:
        argmax_idx = non_zero_rtn['intraDS2'].idxmax()
        rtn_max = non_zero_rtn.loc[argmax_idx, 'rtn_morning']

# 确保 `rtn_max` 不为零后，将显著性因子乘以收益率
if rtn_max != 0:
    merged_data['intraDS2_final'] = merged_data['intraDS2'] * rtn_max
else:
    merged_data['intraDS2_final'] = 0  # 如果找不到有效收益率，设置为 0

# 查看结果
print(merged_data[['intraDS2', 'rtn_morning', 'intraDS2_final']])
```

```python
import numpy as np
import pandas as pd

# 定义因子列，改为实际存在的列名
factor_columns = ['intraDSVol_byT', 'intraDSRtn_byT', 'intraDSStd_byP',
                  'intraDSVol_byP', 'intraDSRtn_byV', 'intraDSStd_byV']

# 初始化组合测试结果数据框
performance_results = pd.DataFrame(index=factor_columns, 
                                   columns=['多空收益率', '多头收益率', '空头收益率', 
                                            '年化波动率', '夏普比率', '最大回撤', '多头换手率'])

# 计算每个因子的多空组合表现
for factor in factor_columns:
    long_stocks = merged_data[merged_data[factor] >= merged_data[factor].quantile(0.7)]
    short_stocks = merged_data[merged_data[factor] <= merged_data[factor].quantile(0.3)]
    
    # 计算多头和空头组合的收益率
    long_rtn = long_stocks['daily_return'].mean()
    short_rtn = short_stocks['daily_return'].mean()
    
    # 多空组合收益率
    long_short_rtn = long_rtn - short_rtn
    
    # 年化波动率
    annual_vol = np.std(merged_data['daily_return']) * np.sqrt(252)  # 假设每年252个交易日
    
    # 夏普比率
    sharpe_ratio = long_short_rtn / annual_vol if annual_vol != 0 else 0
    
    # 最大回撤计算
    merged_data['cumulative_return'] = (1 + merged_data['daily_return']).cumprod()
    peak = merged_data['cumulative_return'].cummax()
    drawdown = (merged_data['cumulative_return'] - peak) / peak
    max_dd = drawdown.min()
    
    # 多头换手率，假设可以根据时间窗口内的股票变动计算
    turnover_rate = long_stocks['stkcd'].nunique() / merged_data['stkcd'].nunique()
    
    # 将结果保存到数据框
    performance_results.loc[factor] = [long_short_rtn, long_rtn, short_rtn, 
                                       annual_vol, sharpe_ratio, max_dd, turnover_rate]

# 输出组合测试结果
print(performance_results)
```

```python
import numpy as np
import pandas as pd

# 创建正交化后的因子
merged_data['intraDSVol_byTN'] = merged_data['intraDSVol_byT'] - merged_data['intraDSVol_byT_20d_mean']
merged_data['intraDSVol_byPN'] = merged_data['intraDSVol_byP'] - merged_data['intraDSVol_byP_20d_mean']
merged_data['intraDS2Rtn_byVN'] = merged_data['intraDSRtn_byV'] - merged_data['intraDSRtn_byV_20d_mean']

# 定义因子
factor_columns = ['intraDSVol_byTN', 'intraDSVol_byPN', 'intraDS2Rtn_byVN']

# 计算多头和空头组合的收益率
results = pd.DataFrame(index=factor_columns, columns=['多空收益率', '多头收益率', '空头收益率', '年化波动率', '夏普比率', '最大回撤', '多头换手率'])

for factor in factor_columns:
    # 获取多头和空头组合
    long_stocks = merged_data[merged_data[factor] >= merged_data[factor].quantile(0.7)]
    short_stocks = merged_data[merged_data[factor] <= merged_data[factor].quantile(0.3)]
    
    # 计算多头和空头的累计收益
    long_return = long_stocks['daily_return'].mean()
    short_return = short_stocks['daily_return'].mean()
    
    # 计算多空收益率
    long_short_return = long_return - short_return
    
    # 计算年化波动率（假设250个交易日）
    annual_volatility = long_stocks['daily_return'].std() * np.sqrt(250)
    
    # 计算夏普比率（假设无风险利率为0）
    sharpe_ratio = long_return / annual_volatility if annual_volatility != 0 else np.nan
    
    # 计算最大回撤
    cumulative_returns = (1 + long_stocks['daily_return']).cumprod()
    max_drawdown = (cumulative_returns / cumulative_returns.cummax() - 1).min()
    
    # 多头换手率（假设换手率等于日间交易量除以总股票数量）
    turnover_rate = (long_stocks['volume_morning'] + long_stocks['volume_afternoon']).sum() / len(long_stocks)
    
    # 将计算结果存储到 DataFrame
    results.loc[factor, '多空收益率'] = long_short_return
    results.loc[factor, '多头收益率'] = long_return
    results.loc[factor, '空头收益率'] = short_return
    results.loc[factor, '年化波动率'] = annual_volatility
    results.loc[factor, '夏普比率'] = sharpe_ratio
    results.loc[factor, '最大回撤'] = max_drawdown
    results.loc[factor, '多头换手率'] = turnover_rate

# 输出结果
print("正交化后的因子组合测试结果")
print(results)
```
